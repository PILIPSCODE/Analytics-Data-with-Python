### Cleaning Data

```python
def clean_data(df):
    df = df[df["PM2.5"] < 300]
    df = df[df["PM10"] < 500]
    df = df[df["O3"] < 270]
    return df

Tiantan_station_df = clean_data(Tiantan_station_df)
Aotizhongxin_station_df = clean_data(Aotizhongxin_station_df)
Changping_station_df = clean_data(Changping_station_df)
Dingling_station_df = clean_data(Dingling_station_df)
Dongsi_station_df = clean_data(Dongsi_station_df)
Gucheng_station_df = clean_data(Gucheng_station_df)
Guanyuan_station_df = clean_data(Guanyuan_station_df)
Huairou_station_df = clean_data(Huairou_station_df)
Shunyi_station_df = clean_data(Shunyi_station_df)
Nongzhanguan_station_df = clean_data(Nongzhanguan_station_df)
Wanliu_station_df = clean_data(Wanliu_station_df)
Wanshouxigong_station_df = clean_data(Wanshouxigong_station_df)
```

```python


def calculate_aqi(AQI):
    if AQI >= 301:
        return "Hazardous"
    elif AQI >= 201:
        return "Very Unhealthy"
    elif AQI >= 151:
        return "Unhealthy"
    elif AQI >= 101:
        return "Unhealthy for Sensitive Groups"
    elif AQI >= 51:
        return "Moderate"
    else:
        return "Good"
    
def calculate_pm25(pm25_value):
    if pm25_value >= 250.4:
        return "Hazardous"
    elif pm25_value >= 150.4:
        return "Very Unhealthy"
    elif pm25_value >= 55.5:
        return "Unhealthy for Sensitive Groups"
    elif pm25_value >= 35.5:
        return "Unhealthy"
    elif pm25_value >= 12.1:
        return "Moderate"
    else:
        return "Good"

def calculate_pm10(pm10_value):
    if pm10_value >= 425.0:
        return "Hazardous"
    elif pm10_value >= 355.0:
        return "Very Unhealthy"
    elif pm10_value >= 255.0:
        return "Unhealthy for Sensitive Groups"
    elif pm10_value >= 155.0:
        return "Unhealthy"
    elif pm10_value >= 55.0:
        return "Moderate"
    else:
        return "Good"




```

hapus semua keterangan waktu ubah menjadi tanggal, selanjutnya buat column AQI dengan mengambil polutan dengan nilai tertinggi, setelah itu gabungkan semua

```python
df_names = ['Aotizhongxin_station_df', 'Changping_station_df', 'Dingling_station_df', 
            'Dongsi_station_df', 'Guanyuan_station_df', 'Gucheng_station_df', 
            'Huairou_station_df', 'Nongzhanguan_station_df', 'Shunyi_station_df', 
            'Tiantan_station_df', 'Wanliu_station_df', 'Wanshouxigong_station_df']


for i,df in enumerate(all_df):
    df['date'] = pd.to_datetime(df[['year', 'month', 'day']])
    df.drop(columns=["No","year","day","hour"])

    #Ambil konsentrasi dengan nilai tertinggi, sesuai standart AQI
    df['AQI'] = df[['PM2.5', 'PM10', 'SO2', 'CO', 'O3','NO2']].max(axis=1)
    exec(f"{df_names[i]} = df")
    
merged_all_df = pd.concat(all_df, ignore_index=True)
```

```python
merged_all_df.head()
```

menggabungkan semua data frame menjadi time frame bulanan

```python
monthly_AQI_df = merged_all_df.resample(rule='ME', on='date').agg({
        "AQI": "median",
        
})
monthly_AQI_df.index = monthly_AQI_df.index.strftime('%B')

AQI_Station = merged_all_df.groupby(by="station")["AQI"].median().sort_values(ascending=True).reset_index()
AQI_Station["category"] = AQI_Station["AQI"].apply(lambda x: calculate_aqi(x))
```

Dari sumber tersebut bulan musim panas dicina pada tahun 2013 - 2017 dimulai dari april - agustus, sedangkan musim dingin oktober - january

```python
def Season(month):
    if month == "April" or month == "May" or month == "June" or month == "July" or month == "August":
        return "Summer"
    elif month == "November" or month == "December" or month == "January":
         return "Winter"
    elif month == "September" or month == "October":
        return "Autumn"
    else:
        return "Spring"
```

### Pertanyaan 2:

Apakah ada perbedaan yang signifikan dalam kualitas udara antara bulan-bulan musim panas dan musim dingin ditahun tersebut?

```python
monthly_Station = monthly_AQI_df.groupby(by="date")["AQI"].median().reset_index()
monthly_Station ["category"] = monthly_Station["AQI"].apply(lambda x: calculate_aqi(x))

month_order = ['January', 'February', 'March', 'April', 'May', 'June', 
               'July', 'August', 'September', 'October', 'November', 'December']

monthly_Station['date'] = pd.Categorical(monthly_Station['date'], categories=month_order, ordered=True)


monthly_Station_sorted = monthly_Station.sort_values('date').reset_index(drop=True)


monthly_Station_sorted["season"] = monthly_Station_sorted["date"].apply(lambda x: Season(x))

monthly_Station_sorted
```

```python
colour = ["#3DC8EF","#099E42","#F8060A","#FD7D00"]



fig = plt.subplots(figsize=(20, 5))
sns.lineplot(x='date', y='AQI', data=monthly_Station_sorted, markerfacecolor='limegreen',linewidth=4,color="black", marker="o",markersize=9)
plt.title("Comparison of Air Quality Between All Season(2013-2017)", loc="center", fontsize=20)


sns.barplot(y="AQI" , x="date",color="#D3D3D3" , hue="season", palette=colour, data=monthly_Station_sorted)
plt.show()
```
